## Reinforcement Learning

In reinforcement learning there is a learning system called an agent in this context, can observe the environment, select and perform actions and get rewards in return (or penalties for negative rewards). It learns by itself what is the best strategy (also called policy) to get the most reward over time. 

The policy defines what action the agent should choose when it is in a given situation.

## Batch and online learning

### Batch learning
In batch learning the system is incapable of learning incrementally i.e. it must be trained using all the avaliable data. This is done offline, after the system is trained and then it is launched into production and runs without learning anymore; it applies what it has learned.

If you want a batch system to learn about new data, you have to train it from scratch on the full dataset (not just the new but also the old data), then replace the old system with the new one.


### Online Learning

In online learning, you train the system incrementally by feeding it data instances sequentially, either individually or by small groups called mini-batches. Each learning step is fast and cheap, so the system can learn about new data on the fly, as it arrives. 

One important parameter of online learning systems is how fast they should adapt to changing data: this is called the learning rate. If you set a high learning rate, then your system will rapidly adapt to new data, but it will also tend to quickly forget the old data.


A big challenge about online learning is that if bad data is fed to the system then the system's performance will decline gradually. To reduce this risk we need to monitor the system closely and promptly switch learning off if you detect a drop in performance. 

## Instance-based versus Model-Based learning

### Instance based learning

The system learns examples by heart, then generalizes to new cases using a similiarity measure.


### Model based learning
Another way to generalize from a set of examples is to build a model of these examples, then use the model to make predictions.


How do you know which values make your model perform the best?
You can define a utility function that measures how good your model is or you can define a cost function that measure how bad it is.


For linear models predictions you want your data as close to the line as possible, the objective is to minimize the distance. 

## Irrelevant Features

Your system will only be capable of learning data if the training data contains enough relevant features and not too many irrelevant ones. A critical part of the success of a Machine Learning project is coming up with a good set of features to train on. Feature engineering involves:

- Feature Selection: Selecting the most useful features to train on among existing features

- Feature extraction: Combining existing features to produce a more useful one

- Creating new features by gathering data



## Countering overtraining

To simplify the model by selecting one with fewer parameters (e.g. a linear model rather an a higher degree polynmial model), by reducing the number of attributes in the training data or by constraining the model

To gather more training data

To reduce the noise in the training data (e.g. fix data errors and remove outliers)

## Regularization

Constraining the model to make it simpler and reduce the risk of overfitting is called regularization


### Hyper-parameter
The amount of regularization to apply during learning can be controlled by a hyper-parameter. A hyper-parameter is a parameter of a learning algorithm.

It must be set prior to training and remains constant during training.

## Countering underfitting

The main options to fix this problem are: 

- Selecting a more powerful model, with more parameters
- Feeding better features to the learning algorithm (feature engineering)
- Reducing the constrains on the model (e.g. reducing the regularization hyper-parameter)


## Testing and Validating

The only way to know how well a model will generalize to new cases is try it on new cases.

Split your data into 2 sets: training set and test set.

Train your model using the training set and test it out using the test set.

The error rate on new cases is called the <b>generalization error</b>

If the training error is low (i.e. your model makes few mistakes on the training set) but the generalization error is high, it means that your model is overfitting the training data.

## Hyperparameter Tuning and Model Selection

If you're hesistant between models (say a linear model and polynomial model). 

You simply holdout part of the training set to evaluate several candidate models and select the best one. The new heldout set is called the <b>validation set.</b>


You train multiple models and various hyperparameters on the reduced training set (i.e. the full training set minus the validation set), and you select the model that performs the model that performs the best on the validation set. 

After this holdout validation process, you train the best model on the full training set (including the validation set) and this gives you the final model. Lastly, you evaluate this final model on the test set to get an estimate of the generalization error.



## Data Mismatch

The most important rule to remember is that the validation set and the test must be representative as possible of the data you expect to use in production. You shuffle the data and make sure half ends up in the test set and half ends up in the validation set making sure no duplicates or near-duplicates end up in both sets.
