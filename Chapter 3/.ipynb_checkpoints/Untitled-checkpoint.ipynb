## Performance Measures

### Confusion Matrix

The general idea is to count the number of times instances of class A are classified as class B. For example, to know the number of times the classifier confused images of 5s with 3s, you would look in the 5th row and 3rd column of the confusion matrix. 


To compute the matrix, get a set of predictions.

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_5, y_train_pred)
```

The first row represents an <b><i>actual class</i></b>, which each column represents <b><i>predicted class</i></b>. The first row of this matrix considers non-5 images <i>(the negative)</i>: 53,057 of them were correctly classified as non-5s (they are called true negatives), while the remaining 1,522 were wrongly classified as 5s <i>(false positives)</i>

The second row considers the images of 5s <i>(the positive class)</i>: 1,325 were wrongly classified as non-5s <i>(false negatives)</i>, while the remaining 4,096 were correctly classified as 5s <i>(true positives)</i>.

### Precision/Recall 


You can use the matrix to create a precision and recall formula which is a better measure for calculating performance. 

precision formula: TP / TP + FP

recall formula: TP / TP + FN 

```python
from sklearn.metrics import precision_score, recall_score

print(f'Precision: {precision_score(y_train_5, y_train_pred)}')


print(f'Recall: {recall_score(y_train_5, y_train_pred)}')
```

As you can see above that the accuracy (precision) is 83%, and the recall is 65% meaning that it only detects 65% of the 5's overall.


You can combine the precision and recall into a single metric called an F1 score. The F1 score is the harmonic mean of the precision and recall. 

The harmonic mean gives much more weight to low values. As a result the classifier will only get a high F1 score if both recall and precision are high.


The formula for F1 score:

F1 = TP / (TP + (FN + FP) / 2)

Computing the F1 score using python: 

```python
from sklearn.metrics import f1_score
f1_score(y_train_5, y_train_pred)
```

The F1 score will favour the values that have similar precision and recall, but in some cases you may care more about the precision (i.e. accuracy) than recall or vice versa.


### Precision/Recall tradeoff

The SGDClassifier makes its classification decisions based on a decision function. Scikit does not let you set the threshold directly, but it does give you access to the decision scores that it uses to make the predictions. 

```python
y_scores = sgd_clf.decision_function([some_digit])
y_scores
```

```python
threshold = 0
y_some_digit_pred = (y_scores > threshold)
print(y_some_digit_pred)
```

```python
threshold = 8000
y_some_digit_pred = (y_scores > threshold)
y_some_digit_pred
```

We know that the image actually represents a 5 but the classifier detects it when the threshold is 0 but not when it's 8000. 

How do you decide which threshold to use? For this you will first need to get the scores of all instances in the training set using the cross_val_predictions but this time specify that you need decision scores instead.

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method="decision_function")
```

Lets create a precision recall curve

```python
from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

Sketch it using matlab

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=16) # Not shown in the book
    plt.xlabel("Threshold", fontsize=16)        # Not shown
    plt.grid(True)                              # Not shown
    plt.axis([-50000, 50000, 0, 1])             # Not shown

plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.show()
```

Plotting precision against recall

```python
recall_90_precision = recalls[np.argmax(precisions >= 0.90)]
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]
```

```python
def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
plt.plot([recall_90_precision, recall_90_precision], [0., 0.9], "r:")
plt.plot([0.0, recall_90_precision], [0.9, 0.9], "r:")
plt.plot([recall_90_precision], [0.9], "ro")
plt.show()
```

Even though we have 99% accuracy, the recall is too low.


### The ROC Curve

The receiver operating characteristic (ROC) curve is a tool used with binary classifiers, but instead of plotting precision vs recall it plot the true positive rate (another name for recall) against the false positive rate. The FPR is the ratio of negative instances that are incorrectly classified as positive. It is equal to one minus the true negative rate, which is the ratio of negative instances that are correctly classified as negative. 

The TNR is called called specificity.

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

Plotting FPR against TPR using Matplotlib.

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0,1], [0,1], 'k--')
    plt.axis([0, 1, 0, 1])                                    # Not shown in the book
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) # Not shown
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    # Not shown
    plt.grid(True)                                            # Not shown

plt.figure(figsize=(8, 6))                                    # Not shown
plot_roc_curve(fpr, tpr)
fpr_90 = fpr[np.argmax(tpr >= recall_90_precision)]           # Not shown
plt.plot([fpr_90, fpr_90], [0., recall_90_precision], "r:")   # Not shown
plt.plot([0.0, fpr_90], [recall_90_precision, recall_90_precision], "r:")  # Not shown
plt.plot([fpr_90], [recall_90_precision], "ro")               # Not shown
plt.show()
```

The higher the recall (TPR), the more false positives (FPR) the classifier produces. The dotted line represents the ROC curve of a purely random classifier; A good classifier stays as far away from that line as possible (toward the top-left corner).

A way to compare classifiers is to measure the area under the curve (AUC). A perfect classifier will have a ROC AUC equal to 1, whereas a purely random classifier will have a ROC AUC equal to 1, whereas a purely random classifier will have a ROC AUC equal to 0.5.




```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5, y_scores)
```

As a rule of thumb you should use the PR curve whenever the positive class is rare or when you care more about the false positives than the false negatives, and the ROC curve otherwise.

### Error Analysis

Let's assume you have found the best model and you want to find ways to improve it. One way to do this is to analyze the types of errors it makes.


First, you can look at the confusion matrix. You need to make predictions using the cross_val_predict() function, then call confusion_matrix() function.

```python
y_train_pred = cross_val_predict(sgd_clf, X_train_scaled, y_train, cv=3)
conf_mx = confusion_matrix(y_train, y_train_pred)
```

```python
conf_mx
```

It's often convenient to look at an image representation of the confusion matrix.

```python
plt.matshow(conf_mx, cmap=plt.cm.gray)
plt.show()
```

The confusion matrix above looks fairly good, since most images are on the diagonal it's means they were classified correctly. The 5's look slightly darker than the other digits, which could mean that there are fewer images of 5s in the dataset or that the classifier does not perform as well on 5s as on the other digits.

```python
row_sums = conf_mx.sum(axis=1, keepdims=True)
```

```python
norm_conf_mx = conf_mx / row_sums
```

If we fill the diagonal with zeros we will only keep the errors.

```python
np.fill_diagonal(norm_conf_mx, 0)
plt.matshow(norm_conf_mx, cmap=plt.cm.gray)
plt.show()
```

Now we can see the kinds of errors the classifer makes. Remember that rows represent actual classes, while the columns represent predicted classes.

The column for class 8 is quite bright that means many images are misclassified as 8. However, the row for class 8 is not that bad. 

Looking at this plot we can see we should spend efforts on reducing the false 8s. For example, you could try to gather more training data for digits that look like 8s (but are not) so the classifier can learn to distinguish them from real 8's. Or you could engineer new features that would help the classifier - for example to count the number of loops in a digit. This way you could determine if it was an 8 or not. You can also try to preprocess images to make some parts stand out more than others.


You can also analyze individual errors.


A big reason for the mistakes is that we used a sgdclassifier, which is a linear model. All it does is assign a weight per class to each pixel, and when it sees a new image it just sums up the weighted pixel intensities to get a score for each class. So 3s and 5s differ only by a few pixels, this model will easily confuse them.

### Multiclass Classification

Each instance was assigned to just one class. In some cases you may want your classifier to output multiple classes for each instance. A classification that outputs multiple binary tags i.e. [1, 0, 1] is called a multilabel classification system.

Let's look at a simple system.


```python
from sklearn.neighbors import KNeighborsClassifier

y_train_large = (y_train >= 7)
y_train_odd = (y_train % 2 == 1)
y_multilabel = np.c_[y_train_large, y_train_odd]

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_multilabel)
```

Now you can make a prediction and notice that it outputs two labels.

```python
knn_clf.predict([some_digit])
```

And it gets it right! The digit 5 is indeed not too large (False) and odd (True)

There are many ways to evaluate a multilabel classifier. For example, one approach is to measure the F1 score for each individual label, then compute the average score. The code below computes the average F1 score across all labels:

```python
y_train_knn_pred = cross_val_predict(knn_clf, X_train, y_multilabel, cv=3)

# y_multilabel is y_true and y_train_knn_pred is the prediction made.
f1_score(y_multilabel, y_train_knn_pred, average="macro")
```

"macro" in the line above means that the score will be calculated for each label, and the result will be the average of those scores. This assumes that all labels are equally important, which may not be the case. If you have more pictures of Alice than of Bob and Charlie, you may want to give more weight to the classifier's score on the picture of Alice.

One option is to simply give each label a weight equal to its support (i.e. the number of instances with that target label)
